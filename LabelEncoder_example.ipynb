```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = ["spam", "ham", "spam", "eggs"]
le.fit(y)

print(le.classes_)            # array(['eggs', 'ham', 'spam'], dtype=object)
print(le.transform(['ham']))  # array([1])
```

```python
print(le.transform(['eggs']))  # array([1])
```
