```python
import pandas as pd
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X = pd.DataFrame({'age': [18, 22, 35, 45, 60, 75], 'income': [20, 30, 50, 80, 40, 100]})
y = [0, 0, 1, 1, 1, 0]

X['age_bin'] = pd.cut(X['age'], bins=[0, 25, 45, 100], labels=['young','adult','senior'])
X['inc_bin'] = pd.qcut(X['income'], q=3, labels=['low','mid','high'])
X_cat = X[['age_bin','inc_bin']]

encoder = OrdinalEncoder()
X_enc = encoder.fit_transform(X_cat)

# use stratify to keep class proportions in train/test
X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.33, random_state=0, stratify=y)

clf = CategoricalNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# avoid the UndefinedMetricWarning by specifying zero_division
print(classification_report(y_test, y_pred, zero_division=0))
```
